```python
import pandas as pd #ενεργοποιώ βιβλιοθήκη Pandas και την μετενομάζω σε pd
df = pd.read_excel("Raisin_Dataset.xlsx") #διαβάζει το excel
```

```python
features = ["Area","MajorAxisLength","MinorAxisLength",
                "Eccentricity","ConvexArea","Extent","Perimeter"]
class_col = ["Class"]
```

```python
X = df[features].values
Y = df["Class"].values
```

```python
print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("Targets:", pd.Series(Y.flatten()).unique())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=12, stratify=Y)
# χωρίζω τα δεδομενα σε 70-30 και ορίζω ενα τυχαιο,
# ορίζω τον τυχαίο διαχωρισμό και τον κάνω αναπαράξιμο
# ορίζω ίδιο αριθμό κατηγοριών για το Υ
```

```python
from sklearn.tree import DecisionTreeClassifier
cart = DecisionTreeClassifier(criterion="entropy", ccp_alpha=0.1, random_state=12) #Ορίζω το μοντέλο Cart και το κριτήριο Gini και βάζω τυχαίο διαχωρισμό
```

# Νέα ενότητα

```python
cart.fit(X_train, Y_train) #Εκπαίδευση του μοντέλου
```

```python
Y_pred = cart.predict(X_test) #Δοκιμή μοντέλου στο τεστ
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test, Y_pred) #ορίζω ως ποσοστό επιτυχιας τη μεταβλητή acc
print(f"Accuracy (70/30): {acc*100}%") #Για να εμφανίσει το ποσοστό επιτυχίας του μοντέλου
print("Depth:", cart.get_depth(), "| Leaves:", cart.get_n_leaves()) #Για να εμφανίσει το μέγεθος του δέντρου και τα φύλλα του
```
